# Evaluation NeurOST SSH-SST: 

This notebook presents the evaluation of the SSH reconstructions based on the NeurOST SSH-SST method presented in [Martin et al., 2024]( https://doi.org/10.1029/2024GL110059) and performed for the **"2021a_SSH_mapping_OSE" ocean data challenge**. 

Note that this method additionally uses L4 SST from the NASA MUR SST product to improve the mapping of mesoscale eddies.

\* This neural network was trained on data from the global mapping to perform global SSH mapping - see 2023a_SSH_mapping_OSE. N.B. the network was re-trained from scratch to ensure 2017 was withheld for this experiment since 2023a_SSH_mapping_OSE has a different withheld validation year.

This method is used to create the [NeurOST NASA PO.DAAC product](https://doi.org/10.5067/NEURO-STV24).
